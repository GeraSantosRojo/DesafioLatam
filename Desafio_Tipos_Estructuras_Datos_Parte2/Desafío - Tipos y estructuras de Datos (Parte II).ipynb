## Crea un DataFrame con el archivo entregado

```python
import pandas as pd

df = pd.read_csv("resumen_resultados_astronautas.csv")
df
```

## Crea la columna "IMC" usando la fórmula entregada. IMC = m / h*h

```python
df['IMC'] = df['peso'] / ((df['altura'] / 100)** 2)
df
```

## Crea la columna "promedio_evaluaciones" con el promedio de las 4 columnas
## de evaluaciones. Si el IMC está entre 18.5 y 24.9, entonces el candidato
## tendrá un IMC adecuado para ser astronauta. Filtra el DataFrame para
## obtener solo aquellas filas donde el IMC esté dentro de los límites
## entregados

```python
df['promedio_evaluaciones'] = df[['evaluacion_fisica', 'evaluacion_psicologica', 'evaluacion_adaptativa', 'evaluacion_medica']].mean(axis=1)
df
```

## Continua... Si el IMC está entre 18.5 y 24.9, entonces el candidato
## tendrá un IMC adecuado para ser astronauta. Filtra el DataFrame para
## obtener solo aquellas filas donde el IMC esté dentro de los límites
## entregados

```python
limite_inferior = 18.5
limite_superior = 24.9
df_filtro = df.loc[(df['IMC'] >= limite_inferior) & (df['IMC'] <= limite_superior)]
df_filtro
```

## El promedio de las 4 evaluaciones debe ser mayor o igual a 87. Filtra el
## DataFrame resultante y agrega esta condición al filtro aplicado en el paso
## anterior.

```python
df_evaluacion = df[df['promedio_evaluaciones'] >= 87]
df_evaluacion
```

## Reinicia el índice del DataFrame resultante

```python
df_evaluacion_reset = df_evaluacion.reset_index(drop=True)
df_evaluacion_reset
```

## Modifica el valor de la columna "califica", para que diga "Si" en todas las filas del
## DataFrame resultante

```python
df_evaluacion_reset['califica'] = 'Si'
df_evaluacion_reset
```

Exporta el DataFrame resultante a un archivo .csv separado por comas con el
nombre "astronautas_calificados.csv"

```python
df_evaluacion_reset.to_csv('astronautas_calificados.csv', index=False, sep=',')
```
