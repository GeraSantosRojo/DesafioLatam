```python
#El nombre de nuestro df actual es muy largo, vamos a simplificarlo.
df_base = df_sales_customer_store_product_team
df_base #confirmamos el cambio.
```

```python
print(df_base.dtypes) #Aquí vemos el tipo de dato que contienen las columnas, para saber si podremos continuar tal cual.
#o bien, será necesario algun cambio.
```

```python
#Creamos nuevas columnas al DF con su respectiva operacion, segun el desafío.
df_base['ProcurementDays'] = df_base['OrderDate'] - df_base['ProcuredDate']
df_base['ShippingDays'] = df_base['ShipDate'] - df_base['OrderDate']
df_base['DeliveryDays'] = df_base['DeliveryDate'] - df_base['ShipDate']
df_base['CustomerDays'] = df_base['ShippingDays'] + df_base['DeliveryDays']

df_base #confirmamos los cambios.
```

```python
print(df_base.dtypes) #confirmamos el typo de dato que tienen nuestras columnas creadas.
```

```python
df_base['CustomerDays'] #Visualizamos la tabla CustomerDays y vemos el tipo de dato, se encuentra en timedelta64.
```

```python
df_base['CustomerDays'] = df_base['CustomerDays'].dt.days #Convertimos la columna a valores numericos.
```

```python
df_base['CustomerDays'].head(5) #confirmamos el cambio
```

```python
#Asignamos bines y labels segun el ejercicio.
bins = [0, 15, 30 , 45, 60, 75, 90]
labels = ['0 to 15 days', '15 to 30 days', '30 to 45 days', '45 to 60 days', '60 to 75 days', '75 to 90 days']

#Creamos columna CustomerDaysInterval con la informacion de bines y labels.
df_base['CustomerDaysInterval'] = pd.cut(df_base['CustomerDays'], bins = bins, labels = labels)
```

```python
df_base.head(20) #Confirmamos cambios
```

```python
#Creamos tabla Pivote.
tabla_pivote = df_base.pivot_table(index = 'Sales Team', columns = 'CustomerDaysInterval', values = 'OrderNumber', aggfunc = 'count', fill_value = 0, observed = False)
tabla_pivote['All'] = tabla_pivote.sum(axis = 1) #Agregamos columna all para sumar las filas.
tabla_pivote.loc['All'] = tabla_pivote.sum() #Agregamos fila All al final para sumar contenido de columnas.
tabla_pivote #Mostramos la tabla.
```

```python
df_base['GrossMargin'] = df_base['Order Quantity'] * (df_base['Unit Price']*(1 - df_base['Discount Applied']) - df_base['Unit Cost'])

df_base.head(2)
```

```python
#Asignamos bines y labels segun el ejercicio.
bins = [0, 100, 1000, 10000, 100000]
    labels = [0.05, 0.10, 0.15, 0.20]

#Creamos columna CustomerDaysInterval con la informacion de bines y labels.
df_base['CommissionsPercentage'] = pd.cut(df_base['GrossMargin'], bins = bins, labels = labels).astype(float)
df_base.head(5)
```

```python
df_base['CommissionsAmount'] = df_base['GrossMargin'] * df_base['CommissionsPercentage']
df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']
df_base.head(5)
```

```python
#pd.options.display.float_format = '{:.2f}'.format  # Formato con dos decimales y separadores de miles

tabla_pivote2 = df_base.pivot_table(
    index = 'Sales Team',
    values = ['GrossMargin', 'NetMargin', 'CommissionsAmount'],
    aggfunc = 'sum', fill_value = 0, observed = False)
tabla_pivote2.loc['All'] = tabla_pivote2.sum() #Agregamos fila All al final para sumar contenido de columnas.
tabla_pivote2_redondeado = tabla_pivote2.round(2)

tabla_pivote2_redondeado
```
