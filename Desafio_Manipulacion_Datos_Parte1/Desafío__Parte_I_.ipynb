## b. Genera una tabla que contenga la unión de las 3 tablas. hint: utiliza sufijos para para las
Columnas que se llaman igual usando el parámetro suffixes de pd.merge().

```python
#Se fusionan las tablas officers e incidents:

merged_df = pd.merge(df_officers, df_incidents, on ='case_number', how = 'inner', suffixes=('_incidents', '_officer'))
```

```python
merged_df.head() #La fusion de tablas df_officers, df_incidents es llamada merged_df
```

```python
#Fusionamos la tercer tabla subjects.

final_df = pd.merge(merged_df, df_subjects, on ='case_number', how = 'inner', suffixes=('', '_subject'))

final_df.head() #La tabla final queda con el nombre final_df
```

## c. Verifica si hay filas duplicadas; si es así, elimínalas

```python
#Verificamos cuatas filas y columnas tiene la tabla.
print(final_df.shape)
```

```python
#Eliminamos tablas duplicadas
df = final_df.drop_duplicates()
#Revisamos si hubo cambios, se mantiene igual.
print(df.shape)
df.head() #Dejamos el nombre de df para mas facilidad.
```

## d. ¿Cuántos sujetos de género F hay en el DataFrame resultante? hint: usa el método .value_counts() sobre la columna.

```python
count_female = df['gender'].value_counts().get('F',0)
print(count_female) #Contamos cuantas filas tienen F en Gender.
```

```python
df_females = df[df['gender'] == 'F']
df_females.head(25) #Aqui podemos confirmar las 20 filas que nos mostro el count.
```

## e. ¿En cuántos números de caso hay por lo menos una sospechosa que sea mujer? hint: utiliza el método unique() para obtener los valores únicos de una columna específica de un DataFrame luego de filtrar.

```python
#Hacemos la variable is_female y filtramos por la columna gender, en el df.
#Donde todo sea F, para que nos muestre los case_numbers
is_female = df[df['gender'] == 'F']['case_number']
is_female.nunique() #.nunique() muestra el numeros de valores unicos.
```

```python
is_female.unique() #unique() muestra la lista de valores unicos.
```

```python
repeated = df[df["gender"] == "F"]["case_number"].value_counts()
print(repeated) #Para complementar la informacion, mostramos cuales son repetidos y cuales no.
```

## f. Genera una tabla pivote que muestre en las filas el género del oficial y en las columnas el género del subject. ¿Cómo interpretas los valores que muestra esta vista?

```python
tabla_pivote = pd.pivot_table(df, index = 'gender', columns = 'gender_subject', aggfunc = 'size',fill_value= 0)
tabla_pivote
```

```python
import pandas as pd
import numpy as np
cleaned_ds = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/Desafío - Manipulación y transformación de datos (Parte I)/Cleaned_DS_Jobs.csv')
```

## b. Utiliza la siguiente lista de valores que serán considerados como nulos:

```python
valores_nulos = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]
```

```python
cleaned_ds.replace(valores_nulos, np.nan, inplace=True)
```

```python
cleaned_ds.head()
print(cleaned_ds.shape)
```

## c. Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna())

```python
cleaned_ds.dropna(inplace = True)
df = cleaned_ds
df #Asignamos la tabla limpia como df, para mayor facilidad.
```

## d. A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo y Máximo. (hint: Utiliza el método apply sobre la columna.)

```python
df[['Salario Estimado Mínimo', 'Salario Estimado Máximo']] = df['Salary Estimate'].apply(lambda x: pd.Series([int(x.split('-')[0].strip()), int(x.split('-')[1].strip())]))
df #Utilizamos metodo apply y lambda.
```

## e. Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint: utilice reemplazo con diccionario usando el método replace sobre la columna.)

```python
df['Size'] = df['Size'].replace(to_replace = [
    '10000+ employees',
    '5001 to 10000 employees',
    '1001 to 5000 employees',
    '201 to 500 employees',
    '51 to 200 employees',
    '501 to 1000 employees',
    'Unknown'], value = [
    'Mega Empresas',
    'Grandes Empresas',
    'Medianas Empresas',
    'Pequeñas Empresas',
    'Pequeñas Grandes Empresas',
    'Microempresas',
    'Empresas sin Información']) #Escribimos por cuales lo cambiamos y escribimos lo que queremos cambiar.

df.head(20) #Confirmamos el cambio
```
